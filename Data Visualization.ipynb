```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# 1. Load your clean data back in
file_path = r'C:\Users\HP\Downloads\4) house Prediction Data Set.csv'
column_names = ['CRIM', 'ZN', 'INDUS', 'CHAS', 'NOX', 'RM', 'AGE', 'DIS', 'RAD', 'TAX', 'PTRATIO', 'B', 'LSTAT', 'MEDV']
df = pd.read_csv(file_path, header=None, names=column_names, sep=r'\s+')
# 2. Set style
sns.set_theme(style="whitegrid")
```

```python
# =================================================================
# OBJECTIVE 1 & 2: CREATE AND CUSTOMIZE PLOTS
# =================================================================
# --- PLOT A: SCATTER PLOT (Relationship) ---
plt.figure(figsize=(8, 5))
sns.scatterplot(data=df, x='RM', y='MEDV', color='purple', alpha=0.7, label='Data Points')
plt.title('House Price vs. Number of Rooms', fontsize=14, fontweight='bold', pad=15)
plt.xlabel('Average Number of Rooms (RM)', fontsize=12)
plt.ylabel('Median Value of Home in $1000s (MEDV)', fontsize=12)
plt.legend(loc='upper left')
plt.tight_layout()
# Objective 3: Export plot as image
plt.savefig('scatter_rooms_vs_price.png', dpi=300)
plt.show()
```

```python
# --- PLOT B: BAR PLOT (Distribution across Categories) ---
# Create an artificial categorical column for data categorization
df['TAX_LEVEL'] = pd.qcut(df['TAX'], q=3, labels=['Low Tax', 'Medium Tax', 'High Tax'])
plt.figure(figsize=(8, 5))
sns.barplot(data=df, x='TAX_LEVEL', y='MEDV', estimator='mean', palette='muted', hue='TAX_LEVEL', legend=False)
plt.title('Average House Price by Property Tax Bracket', fontsize=14, fontweight='bold', pad=15)
plt.xlabel('Tax Bracket', fontsize=12)
plt.ylabel('Average Price ($1000s)', fontsize=12)
plt.tight_layout()
# Export plot as image
plt.savefig('bar_tax_vs_price.png', dpi=300)
plt.show()
```

```python
# --- PLOT C: LINE CHART (Trends) ---
# Sort data by AGE to track trend line progression
df_sorted = df.sort_values(by='AGE')

plt.figure(figsize=(10, 5))
plt.plot(df_sorted['AGE'], df_sorted['NOX'], color='teal', linewidth=2, label='Nitric Oxide Level')
plt.title('Environmental Pollution (NOX) Trend by Age of Home', fontsize=14, fontweight='bold', pad=15)
plt.xlabel('Proportion of Units Built Prior to 1940 (AGE)', fontsize=12)
plt.ylabel('Nitric Oxides Concentration', fontsize=12)
plt.legend(loc='upper left')
plt.tight_layout()
# Export plot as image
plt.savefig('line_age_vs_pollution.png', dpi=300)
plt.show()
```
