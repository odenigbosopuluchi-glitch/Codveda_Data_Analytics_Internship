```python
import pandas as pd
# 1. Define the correct column names for the Boston Housing dataset
column_names = [
    'CRIM', 'ZN', 'INDUS', 'CHAS', 'NOX', 'RM', 'AGE', 
    'DIS', 'RAD', 'TAX', 'PTRATIO', 'B', 'LSTAT', 'MEDV'
]

# 2. File path
file_path = r'C:\Users\HP\Downloads\4) house Prediction Data Set.csv'

# 3. Load dataset with correct separator and column headers
df = pd.read_csv(file_path, header=None, names=column_names, sep=r'\s+')

# 4. Clean summary output
print("✨ Dataset loaded with all columns correctly separated!")
print(f"📊 Total Rows: {df.shape[0]} | Total Columns: {df.shape[1]}")
print("-" * 50)

# Display the first few rows with perfect headers
df.head()
```

```python
# =================================================================
# OBJECTIVE 1: REMOVE DUPLICATE ROWS
# =================================================================
print("--- 1. DUPLICATE CHECK ---")
initial_count = len(df)

# Check and drop duplicate rows
df = df.drop_duplicates()
final_count = len(df)
removed_duplicates = initial_count - final_count

print(f"✅ Removed {removed_duplicates} duplicate rows.")
print(f"📊 Total rows remaining: {final_count}")
print("-" * 50)
```

```python
# =================================================================
# OBJECTIVE 2: IDENTIFY AND HANDLE MISSING VALUES
# =================================================================
print("--- 2. MISSING VALUES CHECK ---")
missing_count = df.isnull().sum().sum()

if missing_count > 0:
    print(f"⚠️ Found {missing_count} missing values. Handling with median imputation...")
    # Fill missing values with the median of each column
    df = df.fillna(df.median())
    print("✅ Missing values successfully handled!")
else:
    print("✅ Perfect! No missing values detected in the dataset.")
print("-" * 50)
```

```python
# =========================================================
# OBJECTIVE 2: IDENTIFY AND HANDLE MISSING VALUES
# 
print("\n--- 2. MISSING VALUES CHECK ---")
# Count total missing cells in the dataset
total_missing = df.isnull().sum().sum()
print(f"Total missing values found: {total_missing}")

# Handle missing entries securely using median imputation
df = df.fillna(df.median())
print("✅ Missing value check complete. Any null fields handled using column medians.")
print("-" * 50)
```

```python

# =================================================================
# OBJECTIVE 3: STANDARDIZE DATA FORMATS & SAVE
# =================================================================
print("--- 3. STANDARDIZATION & EXPORT ---")

# Ensure all data points are properly formatted as floating numbers
df = df.astype(float)
print("✅ All columns standardized to floating-point numerical values.")

# Export clean dataset to a new CSV file
df.to_csv("cleaned_house_prediction.csv", index=False)
print("💾 Cleaned dataset saved successfully as 'cleaned_house_prediction.csv'!")
```
